### Variables numéricas

Podemos comprobar que las variables que contienen valores atípicos son: 'Magnesium', 'Alcaliny of ash', 'Proanthocyanins', 'Color intensity', 'Hue' y 'Ash'.

Voy a crear dos grupos, uno con los outliers y el otro no outliers. Después voy a transformar las variables de los valores atípicos utilizando la estandarización de RobustScaler y para los datos no atípicos utilizaré escalado minmax para reescalar los datos a un rango específico, como 0 y 1, preservar la relación entre los datos originales y mantener la distribución relativa de los datos.  

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import RobustScaler, MinMaxScaler

# Grupo variables valores atípicos
numeric_features = ['Magnesium', 'Alcaliny of ash', 'Proanthocyanins', 'Color intensity', 'Hue', 'Ash']
numeric_transformer = Pipeline(
    [
        ('imputer_num', SimpleImputer(strategy = 'median')),
        ('robust', RobustScaler())
    ]
)
# Grupo variables valores no atípicos
numeric_features_minmax = ['Classe', 'Alcohol', 'Malic acid', 'Total phenols', 'Flavanoids', 'Nonflavanoid phenols', 'OD280/OD315 of diluted wines', 'Proline']
numeric_transformer_minmax = Pipeline(
    [
        ('imputer_num', SimpleImputer(strategy = 'median')),
        ('minmax', MinMaxScaler())
    ]
)
```

PIPELINE

Ya que tenemos creadas todas las pipelines, utilizamos ColumnTransformer para utilizar a cada una de las variables. 

```python
from sklearn.compose import ColumnTransformer

preprocessor = ColumnTransformer(
    [
        
        ('numericals', numeric_transformer, numeric_features),
        ('numericals_minmax', numeric_transformer_minmax, numeric_features_minmax) 
    ],
    remainder = 'drop'
)
# con remainder eliminamos las columnas que no estan especificas
```

Como la variable dependiente va a ser 'Classe', para las variables independientes voy a establecer el resto de variables.

Ahora voy a dividir los datos en train y test

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

- El primer algoritmo será el del árbol de decisión, que imita facilmente el pensamiento a nivel humano. 

```python
from sklearn.tree import DecisionTreeClassifier

# Creo decisión del árbol de classificación
clf = DecisionTreeClassifier(random_state=42)
## Pipeline
pipeline = Pipeline ([
            ('preprocessing', preprocessor),
            ('model', clf)
        ])
```

```python
# Entreno 
clf = clf.fit(x_train,y_train)
```

```python
# Ahora obtengo resultados
r_sq = clf.score(x_train, y_train)
print(f"Coeficiente de determinación: {r_sq}")
```

##### Como se puede comprobar, el modelo se ajusta pefectamente a los datos, ya que el valor del R2 es el máximo. :)

```python
from sklearn.model_selection import GridSearchCV

clf = DecisionTreeClassifier()
# Define los parámetros que quieres probar
param_grid = {
    'max_depth': [3, 5, 7],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'criterion': ['gini', 'entropy']
}

# Búsqueda de hiperparámetros utilizando GridSearchCV
grid_search_clf = GridSearchCV(estimator=clf, param_grid=param_grid, cv=5)
grid_search_clf.fit(x_train, y_train)

# Predicción en el conjunto de prueba
predictions = grid_search_clf.predict(x_test)

# Mejor combinación de parámetros
best_params_clf = grid_search_clf.best_params_
print("Mejores parámetros:", best_params_clf)

# Evaluar el modelo con los mejores parámetros
best_model = grid_search_clf.best_estimator_
accuracy = best_model.score(x_test, predictions)
print("Exactitud del mejor modelo en datos de prueba:", accuracy)
```
